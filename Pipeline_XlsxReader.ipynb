```python
import pandas as pd

# --- Đọc file Excel ---
df = pd.read_excel("TrainData.xlsx")

# --- Chỉ lấy 3 cột cần thiết ---
df = df[["ID", "article", "summary"]]

# --- Xuất ra JSON (một mảng object) ---
df.to_json("TrainData.json", orient="records", force_ascii=False, indent=2)

# --- Xuất ra JSONL (mỗi dòng là 1 JSON object) ---
df.to_json("TrainData.jsonl", orient="records", lines=True, force_ascii=False)
```
