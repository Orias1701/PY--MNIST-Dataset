## 1. Cài đặt và import

````python
import sys
sys.path.insert(0, '.')

from lib import (
    MNISTDataLoader,
    MNISTClassifier,
    plot_samples,
    plot_training_history,
    plot_confusion_matrix,
    print_classification_report,
)
import matplotlib.pyplot as plt
````

## 2. Tải dữ liệu MNIST

````python
loader = MNISTDataLoader(normalize=True, flatten=True)
X_train, y_train, X_test, y_test = loader.load()

print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"X_test: {X_test.shape}, y_test: {y_test.shape}")
print(f"Input shape (một mẫu): {loader.input_shape}")
````

## 3. Xem mẫu ảnh

````python
fig = plot_samples(X_train, y_train, n_rows=2, n_cols=5, title="MNIST – 10 mẫu")
plt.show()
````

## 4. Huấn luyện mô hình

````python
classifier = MNISTClassifier(model_type="logistic", random_state=42)
classifier.fit(X_train, y_train, X_val=X_test, y_val=y_test)

print(f"Accuracy trên train: {classifier.score(X_train, y_train):.4f}")
print(f"Accuracy trên test:  {classifier.score(X_test, y_test):.4f}")
````

## 5. Dự đoán và đánh giá

````python
y_pred = classifier.predict(X_test)
class_names = loader.get_class_names()

print("Báo cáo phân lớp:")
print_classification_report(y_test, y_pred, class_names=class_names)
````

## 6. Ma trận nhầm lẫn

````python
fig = plot_confusion_matrix(y_test, y_pred, class_names=class_names, title="Confusion Matrix – MNIST")
plt.show()
````

## 7. (Tùy chọn) Lịch sử huấn luyện

Với mô hình sklearn một bước, history chỉ có 1 điểm; hàm vẽ vẫn hoạt động.

````python
if classifier.history.get("accuracy"):
    fig = plot_training_history(classifier.history, title="Training history")
    plt.show()
````
